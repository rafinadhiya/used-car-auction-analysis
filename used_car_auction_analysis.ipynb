## Background

---

The process of auctioning used cars is crucial in vehicle inventory management to prevent stockpiling and ensure profitability for rental companies, leasing firms, dealerships, automotive manufacturers, and fleet managers. **A primary challenge in auctions is selling vehicles at prices close to their wholesale market value to maximize profit and minimize loss.**

To assist sellers in setting fair prices, the Manheim Market Report (MMR) is utilized as a wholesale pricing guide that provides estimated market values for vehicles. MMR pricing calculations are based on millions of sales transactions from the previous 13 months. ([Manheim Valuation](https://site.manheim.com/en/services/valuation.html))

Despite referencing MMR, not all vehicles sell at or above market value, which can impact profitability. Various critical factors, such as condition grade, brand, age, odometer reading, production year, promotions, sale location, and seller performance, influence auction outcomes. **This analysis aims to identify key factors that can enhance auction profitability.**

# Business Metrics

The following metrics are used to evaluate the effectiveness of used car sales, focusing on three key metrics: **revenue**, **MMR retention rate**, and the **price difference between the average selling price and MMR**.

| **Metric**                  | **Definition**                                                          | **Formula**                                                  | **Usage**                                                                 |
|-----------------------------|------------------------------------------------------------------------|-------------------------------------------------------------|---------------------------------------------------------------------------|
| **Revenue**                 | Total revenue from used car sales.                                     | Revenue = Σ(Selling Price)                                  | Measures the total value of sales to evaluate overall performance.       |
| **MMR Retention Rate**      | Measures how close the selling price is to the wholesale market value (MMR). | MMR Retention Rate = (Selling Price ÷ MMR Value) × 100%     | Determines whether cars are sold close to or above market value.         |
| **Price Difference from MMR** | The difference between the average selling price and the MMR value.   | Avg. Price Difference = Avg. Selling Price - Avg. MMR Value | Identifies whether sales are approaching or deviating from market values. |

**Explanation of Business Metrics:**

- **Total Revenue** is calculated from the accumulated selling prices of all vehicles successfully sold within a specific period.  
- **MMR Retention Rate** measures whether the selling price of vehicles is close to or exceeds the MMR, which serves as a key benchmark in wholesale market valuation.  
- **Price Difference from MMR** helps assess how well the sales performance compares to the market value.

**Next Steps**

Unique values will be calculated for the combination of `brand` and `model`.

```python
# Calculate the frequency of occurrences for the combination of 'brand' and 'model' where 'bodytype' values are missing
df_missing_type[['brand', 'model']].value_counts()
```

**Analysis**

- The first step in addressing NaN values in the `bodytype` column is to identify which combinations of `brand` and `model` have missing `bodytype` values.  

- This process helps to identify patterns in the missing data.  

```python
# Fill missing 'bodytype' values using the mode of 'bodytype' grouped by 'brand' and 'model'
data['bodytype'] = data.groupby(['brand', 'model'])['bodytype'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan))

# If there are still missing values after the groupby operation, fill them with the overall mode
data['bodytype'] = data['bodytype'].fillna(data['bodytype'].mode()[0])
```

**Handling**

**Reason for Handling `bodytype` by Group**

- The goal is to retain specific information for each group (`brand` and `model`). If `groupby` is not applied and NaN values are filled with the mode of the entire dataset, important information about the distinct characteristics of each `brand` and `model` may be lost.

- Each group of `brand` and `model` might exhibit unique patterns related to `bodytype`. Filling NaN values based on groupby is an effort to preserve those patterns.  

```python
# Count the number of NaN (missing values) in each column
data.isna().sum()
```

**Handling**

Successfully handled missing values in the `bodytype` column.

```python
# Creating the mapping dictionaries

# 'day_mapping' dictionary that maps the abbreviated day names (e.g., Sun) to their full names (e.g., Sunday)
day_mapping = {
    'Sun': 'Sunday',
    'Mon': 'Monday',
    'Tue': 'Tuesday',
    'Wed': 'Wednesday',
    'Thu': 'Thursday',
    'Fri': 'Friday',
    'Sat': 'Saturday'
}

# 'month_mapping' dictionary that maps the abbreviated month names (e.g., Jan) to their full names (e.g., January)
month_mapping = {
    'Jan': 'January',
    'Feb': 'February',
    'Mar': 'March',
    'Apr': 'April',
    'May': 'May',
    'Jun': 'June',
    'Jul': 'July',
    'Aug': 'August',
    'Sep': 'September',
    'Oct': 'October',
    'Nov': 'November',
    'Dec': 'December'
}

# Replacing the values in the 'day' column using the 'day_mapping' dictionary
data['day'] = data['day'].replace(day_mapping)
# Replacing the values in the 'month' column using the 'month_mapping' dictionary
data['month'] = data['month'].replace(month_mapping)

# Displaying a sample of 5 rows from the modified columns
data[['saledate', 'day', 'month', 'date', 'hour']].sample(5)
```

```python
# Defining a function to determine the quarter based on the month name
def get_quarter(month):
  # Condition to determine the first quarter
  if month in ['January', 'February', 'March']:
    return 1
  # Condition to determine the second quarter
  elif month in ['April', 'May', 'June']:
    return 2
  # Condition to determine the third quarter
  elif month in ['July', 'August', 'September']:
    return 3
  # Condition to determine the fourth quarter
  elif month in ['October', 'November', 'December']:
    return 4
  # Condition if the month name is invalid
  else:
    return None

# Adding a 'quarter' column to the DataFrame 'data'
data['quarter'] = data['month'].apply(get_quarter)

# Displaying a sample of 7 rows from the modified columns
data[['saledate', 'day', 'month', 'quarter', 'date', 'hour']].sample(7)
```

Successfully created and displayed the new `quarter` column.

```python
# Defining the function categorize_day(day) to categorize the day as 'Weekend' or 'Weekday'
def categorize_day(day):
  # Condition to determine if the day is a 'Weekend'
  if day in ['Saturday', 'Sunday']:
    return 'Weekend'
  # Condition to determine if the day is a 'Weekday'
  else:
    return 'Weekday'

# Adding a 'day_category' column to the DataFrame 'data'
data['day_category'] = data['day'].apply(categorize_day)

# Displaying a random sample of rows with the modified columns
data[['saledate', 'day', 'day_category', 'month', 'quarter', 'date', 'hour']].sample(15)
```

## 3. Is there a significant difference in selling prices among the 5-grade car scale?

Performs an ANOVA test to determine if there are significant differences in selling prices among the 5-scale car grades. **( Statistical Test: ANOVA ).**

```python
# Group data by grade and store in separate variables.
grade_1 = data['sellingprice'][data['grade'] == 1]
grade_2 = data['sellingprice'][data['grade'] == 2]
grade_3 = data['sellingprice'][data['grade'] == 3]
grade_4 = data['sellingprice'][data['grade'] == 4]
grade_5 = data['sellingprice'][data['grade'] == 5]
```

Filter the data in the DataFrame data to get subgroup data that only contains the selling prices of cars with each grade.

**H0: There is no significant difference in selling price among the 5 car scale grades**

**H1: There is a significant difference in selling price among the 5 car scale grades**

```python
# One-way ANOVA test
res = stats.f_oneway(grade_1, grade_2, grade_3, grade_4, grade_5)

# Print p-value
print('P-value:', res.pvalue)
```

The p-value from the ANOVA test was less than 0.05. This indicates that the **null hypothesis (H0) was rejected and the alternative hypothesis (H1) was accepted.** There is a statistically significant difference in selling prices among the different car grade groups.

## 8. Is there a significant difference in selling prices across sales locations?

Performs ANOVA and Kruskal-Wallis tests to determine if the price differences between locations are statistically significant. **( Statistical Tests: ANOVA and Kruskal-Wallis ).**

**Null Hypothesis (H0): There is no statistically significant difference in selling prices across different sales locations.**

**Alternative Hypothesis (H1): There is a statistically significant difference in selling prices across different sales locations.**

```python
# Group selling prices by state
grouped_prices = [group["sellingprice"].values for name, group in data.groupby("state")]

# Perform ANOVA test
anova_result = f_oneway(*grouped_prices)
anova_result
```

```python
# Kruskal-Wallis test
kruskal_result = kruskal(*grouped_prices)
kruskal_result
```

**ANOVA test** assumes the data follows a normal distribution, while **Kruskal-Wallis test** is a non-parametric test that does not depend on distribution assumptions. **Kruskal-Wallis test is performed with the aim of verifying the analysis results.**

**Result Interpretation:**

- p-value of 0.0 (less than 0.05) in both tests indicates that H0 is successfully rejected and H1 is accepted. There is a significant difference in selling price at each sales location.
- Although the ANOVA and Kruskal-Wallis tests indicate a significant difference in selling price between states, this difference is not practically significant (based on descriptive statistical analysis). ## 11. Do cars with low mileage and good condition always sell for higher prices?

Conducts Pearson and Spearman correlation tests to examine the relationship between mileage (odometer), grade, and selling price **( Statistical Tests: Pearson and Spearman Correlation )**

```python
# Check if the data is normally distributed between odometer and selling price
correlation, p_value = pearsonr(data['odometer'], data['sellingprice'])
print(f"Pearson correlation between odometer and selling price: {correlation:.2f}, p-value: {p_value:.3f}")

# Interpretation
if p_value < 0.05:
    print("Reject H0. There is a correlation between odometer and selling price")
else:
    print("Accept H0. There is no correlation between odometer and selling price")

# Check whether the data is normally distributed
# Odometer Histogram
plt.figure(figsize=(8, 6))
plt.hist(data['odometer'], bins=20, color='lightpink', edgecolor='black')  # Using a soft pastel color
plt.title('Odometer Histogram')
plt.xlabel('Odometer')
plt.ylabel('Frequency')
plt.show()

# Selling price histogram
plt.figure(figsize=(8, 6))
plt.hist(data['sellingprice'], bins=20, color='lightyellow', edgecolor='black')  # Using a soft pastel color
plt.title('Selling Price Histogram')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()

# Shapiro-Wilk test for odometer and selling price
shapiro_odometer = stats.shapiro(data['odometer'])
shapiro_sellingprice = stats.shapiro(data['sellingprice'])

print(f"Shapiro-Wilk test for odometer: Statistic={shapiro_odometer.statistic:.3f}, p-value={shapiro_odometer.pvalue:.3f}")
print(f"Shapiro-Wilk test for selling price: Statistic={shapiro_sellingprice.statistic:.3f}, p-value={shapiro_sellingprice.pvalue:.3f}")

# If p-value < 0.05, then reject the hypothesis that the data is normally distributed
if shapiro_odometer.pvalue < 0.05:
    print("Odometer data is not normally distributed")
else:
    print("Odometer data is normally distributed")

if shapiro_sellingprice.pvalue < 0.05:
    print("Selling price data is not normally distributed")
else:
    print("Selling price data is normally distributed")

# If the data is not normally distributed, use Spearman correlation
if shapiro_odometer.pvalue < 0.05 or shapiro_sellingprice.pvalue < 0.05:
    correlation, p_value = spearmanr(data['odometer'], data['sellingprice'])
    print(f"Spearman correlation between odometer and selling price: {correlation:.2f}, p-value: {p_value:.3f}")
    if p_value < 0.05:
        print("Reject H0. There is a correlation between odometer and selling price")
    else:
        print("Accept H0. There is no correlation between odometer and selling price")
```

```python
# Check if the data is normally distributed between grade and selling price
correlation, p_value = pearsonr(data['grade'], data['sellingprice'])
print(f"Pearson correlation between grade and selling price: {correlation:.2f}, p-value: {p_value:.3f}")

# Interpretation
if p_value < 0.05:
    print("Reject H0. There is a correlation between grade and selling price")
else:
    print("Accept H0. There is no correlation between grade and selling price")

# Check whether the data is normally distributed
# Grade Histogram
plt.figure(figsize=(8, 6))
plt.hist(data['grade'], bins=20, color='lightpink', edgecolor='black')  # Using a soft pastel color
plt.title('Grade Histogram')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()

# Selling price histogram
plt.figure(figsize=(8, 6))
plt.hist(data['sellingprice'], bins=20, color='lightblue', edgecolor='black')  # Using a soft pastel color
plt.title('Selling Price Histogram')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()

# Shapiro-Wilk test for grade and selling price
shapiro_grade = stats.shapiro(data['grade'])
shapiro_sellingprice = stats.shapiro(data['sellingprice'])

print(f"Shapiro-Wilk test for grade: Statistic={shapiro_grade.statistic:.3f}, p-value={shapiro_grade.pvalue:.3f}")
print(f"Shapiro-Wilk test for selling price: Statistic={shapiro_sellingprice.statistic:.3f}, p-value={shapiro_sellingprice.pvalue:.3f}")

# If p-value < 0.05, then reject the hypothesis that the data is normally distributed
if shapiro_grade.pvalue < 0.05:
    print("Grade data is not normally distributed")
else:
    print("Grade data is normally distributed")

if shapiro_sellingprice.pvalue < 0.05:
    print("Selling price data is not normally distributed")
else:
    print("Selling price data is normally distributed")

# If the data is not normally distributed, use Spearman correlation
if shapiro_grade.pvalue < 0.05 or shapiro_sellingprice.pvalue < 0.05:
    correlation, p_value = spearmanr(data['grade'], data['sellingprice'])
    print(f"Spearman correlation between grade and selling price: {correlation:.2f}, p-value: {p_value:.3f}")
    if p_value < 0.05:
        print("Reject H0. There is a correlation between grade and selling price")
    else:
        print("Accept H0. There is no correlation between grade and selling price")
```

The data is not normally distributed, so a Spearman correlation test will be conducted to see the relationship between `odometer` and `grade` with `selling price` **( Inferential Statistical Analysis ).**

**H0: There is no correlation between odometer and selling price.**

**H1: There is a correlation between odometer and selling price.**

**H0: There is no correlation between grade and selling price.**

**H1: There is a correlation between grade and selling price.**

```python
# Correlation between Odometer and Selling Price
spearman_corr_odometer, spearman_p_odometer = spearmanr(data['odometer'], data['sellingprice'])

# Correlation between Grade and Selling Price
spearman_corr_grade, spearman_p_grade = spearmanr(data['grade'], data['sellingprice'])

# Display the correlation results
print(f"Spearman Odometer: rho = {spearman_corr_odometer:.2f}, p-value = {spearman_p_odometer:.4f}")
print(f"Spearman Grade: rho = {spearman_corr_grade:.2f}, p-value = {spearman_p_grade:.4f}")
```

A p-value less than 0.05 indicates that the null hypothesis (H0) is rejected, and the alternative hypothesis (H1) is accepted. This suggests that there is a correlation between odometer and selling price, as well as a correlation between grade and selling price.

**Conclusion:**  
- **Odometer:**  
  - **Strong negative correlation** (rho = -0.70) → **The higher the mileage, the lower the selling price**.  
  - **Statistically significant** (p-value = 0.0000).  

- **Grade (Condition):**  
  - **Moderate positive correlation** (rho = 0.44) → **Cars in better condition (higher grade) tend to sell for a higher price**.  
  - **Statistically significant** (p-value = 0.0000).  

Thus, cars with **lower mileage and better condition (higher grade)** indeed tend to **sell at higher prices**.
